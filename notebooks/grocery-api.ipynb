```python
import pandas as pd
import googlemaps
```

This notebook contains an initial attempt at querying the Google Maps API for grocery stores

```python
# API key needs to be added to .env file
API_KEY = "GOOGLE_MAPS_API_KEY"
```

```python
gmaps = googlemaps.Client(key=API_KEY)
```

```python
location = (40.7128, -74.0060)

# Define the search query
query = 'grocery stores'

# Use the Google Maps Places API to search for the query
result = gmaps.places(query, location=location)
```
